**Описание предметной области**

Небольшой книжный интернет-магазин хранит информацию о книгах, их авторах, покупателях и совершённых заказах. Необходимо создать базу данных, отражающую эти сущности, и выполнить ряд запросов для получения аналитической информации.

Структура базы данных
База данных должна состоять из четырёх таблиц:

**authors** – информация об авторах:

`author_id` – целое число, первичный ключ

`name` – текст, имя автора

`country` – текст, страна проживания

**books** – информация о книгах:

`book_id` – целое число, первичный ключ

`title` – текст, название книги

`author_id` – целое число, внешний ключ к таблице authors

`price` – вещественное число, цена книги

`published_year` – целое число, год издания

**customers** – информация о покупателях:

`customer_id` – целое число, первичный ключ

`name` – текст, имя покупателя

`email` – текст, электронная почта

**orders** – информация о заказах:

`order_id` – целое число, первичный ключ

`customer_id` – целое число, внешний ключ к таблице customers

`book_id` – целое число, внешний ключ к таблице books

`quantity` – целое число, количество экземпляров

`order_date` – текст, дата заказа в формате YYYY-MM-DD

**Часть 2.** Написание SQL-запросов

**Простой JOIN**

Выведите список всех книг с указанием имени автора. Результат должен содержать: название книги, имя автора, цену, год издания.

```python
cursor.execute("""
SELECT books.title, authors.name, books.price, books.published_year
FROM books
INNER JOIN authors ON books.author_id = authors.author_id
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
print(f"Got {len(rows)} rows")
```

**Агрегация с группировкой**

Для каждого автора посчитайте общее количество проданных экземпляров его книг. Выведите имя автора и суммарное количество продаж. Отсортируйте по убыванию количества.

```python
cursor.execute("""
SELECT authors.name,
    SUM(orders.quantity) AS authors_sold_quantity
FROM authors
LEFT JOIN books ON authors.author_id = books.author_id
LEFT JOIN orders ON books.book_id = orders.book_id
GROUP BY authors.name
ORDER BY authors_sold_quantity DESC;
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
print(f"Got {len(rows)} rows")
```

**Топ самых продаваемых книг**

Найдите топ-3 книг по суммарному количеству проданных экземпляров. Выведите название книги, имя автора и общее количество продаж.

```python
cursor.execute("""
SELECT books.title, authors.name, SUM(orders.quantity) AS title_sold_quantity
FROM books
LEFT JOIN authors ON books.author_id = authors.author_id
LEFT JOIN orders ON books.book_id = orders.book_id
GROUP BY books.title
ORDER BY title_sold_quantity DESC
LIMIT 3
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
print(f"Got {len(rows)} rows")
```

**Клиенты с высокой суммой заказов**

Найдите покупателей, у которых общая сумма всех заказов превышает 50 (условная единица). Сумма заказа вычисляется как price * quantity. Выведите имя покупателя, его email и общую сумму. Отсортируйте по сумме по убыванию.

```python
cursor.execute("""
SELECT customers.name, customers.email, SUM(books.price * orders.quantity) AS margin
FROM customers
LEFT JOIN orders ON customers.customer_id = orders.customer_id
LEFT JOIN books ON orders.book_id = books.book_id
GROUP BY customers.name
HAVING margin > 50
ORDER BY margin DESC
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
print(f"Got {len(rows)} rows")
```

**Средняя цена книг по годам издания**

Для каждого года издания (published_year) посчитайте среднюю цену книг, выпущенных в этом году. Выведите год и среднюю цену, округлённую до двух знаков после запятой.

```python
cursor.execute("""
SELECT books.published_year, ROUND(AVG(books.price), 2)
FROM books
GROUP BY published_year
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
print(f"Got {len(rows)} rows")
```

**Книги дороже средней цены**

Выведите книги, цена которых выше средней цены всех книг в базе. Покажите название, автора, цену и год издания.

```python
cursor.execute("SELECT ROUND(AVG(price), 2) FROM books")
avg_price = cursor.fetchone()[0]
print(f"Средняя цена всех книг: ${avg_price}")
print("\nКниги с ценой выше среднего:")
cursor.execute("""
SELECT books.title, authors.name, books.price, books.published_year
FROM books
INNER JOIN authors ON books.author_id = authors.author_id
WHERE books.price > (SELECT AVG(price) FROM books)
ORDER BY books.price DESC
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
print(f"Got {len(rows)} rows")
```

**Автор с максимальной выручкой**

Определите автора, чьи книги принесли наибольшую суммарную выручку (сумма price * quantity по всем заказам). Выведите имя автора и общую выручку.

```python
cursor.execute("""
SELECT authors.name, ROUND(SUM(books.price * orders.quantity), 2) AS authors_margin
FROM authors
INNER JOIN books ON authors.author_id = books.author_id
INNER JOIN orders ON books.book_id = orders.book_id
GROUP BY authors.name
ORDER BY authors_margin DESC
LIMIT 1
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
print(f"Got {len(rows)} rows")
```

**Детали заказов с дополнительной информацией**

Для каждого заказа выведите: дату заказа, имя покупателя, название книги, имя автора, количество, цену одной книги и общую стоимость позиции (количество * цена). Отсортируйте по дате (от новых к старым) и по покупателю.

```python
cursor.execute("""
SELECT  orders.order_date, customers.name, books.title, authors.name, orders.quantity, books.price,
orders.quantity * books.price
FROM orders
INNER JOIN customers ON orders.customer_id = customers.customer_id
INNER JOIN books ON orders.book_id = books.book_id
INNER JOIN authors ON books.author_id = authors.author_id
ORDER BY customers.name ASC, orders.order_date DESC
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
print(f"Got {len(rows)} rows")
```
